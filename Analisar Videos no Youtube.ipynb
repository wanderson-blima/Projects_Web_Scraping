## Projeto: Analisar Videos no Youtube

#### O programa tem como objetivo localizar e armazenar videos do youtube com o assunto de sua prefer√™ncia.

```python
# Instalando a biblioteca selenium caso necess√°rio:
# !pip install selenium
```

```python
# Importando os m√©todos do selenium necess√°rios para o programa:
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

# H√°bilitando o drive do chrome: √© necess√°rio baixar o arquivo do drive e coloca-lo dentro da pasta rais do an√°conda;
## Link para dawnload do drive: Chrome -> https://sites.google.com/chromium.org/driver/
driver = webdriver.Chrome()   
```

```python
# Aqui acessamos o Site do Youtube e pesquisamos videos sobre um assunto de sua preferencia: 
pesquisa_name = input('Qual assunto de sua prefer√™ncia gostaria de pesquisar? ')
driver.get(f"https://www.youtube.com/results?search_query={pesquisa_name}")

# √â utilizamos Javascript dentro de um loop para dar "scroll" na p√°gina e localizar mais vid√©os:
for i in range(10):
    scroll = 4000 * i
    driver.execute_script(f'window.scroll(0, {scroll});')
    time.sleep(2)

# Armazenando os elementos dos videos em uma variavel:
videos = driver.find_elements(By.ID, 'video-title')

# Diz qual o titulo do site:
print(driver.title)

# Fazendo um loop para printar cada video e seu titulo individualmente:
quantidade_videos = 0

for video in videos:
    video_link = video.get_attribute('href')
    video_titulo = video.get_attribute('title')
    if video_link != None: 
        print(f'{video_link} : {video_titulo}')
        quantidade_videos += 1
```

```python
# Diz quantos videos foram localizados com o assunto escolhido:
print(f'Foram localizados {quantidade_videos} videos sobre {pesquisa_name}!')
```
